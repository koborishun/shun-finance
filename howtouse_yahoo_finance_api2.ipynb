```python
import sys
from yahoo_finance_api2 import share
from yahoo_finance_api2.exceptions import YahooFinanceError
import pandas as pd
import datetime 
 
my_share = share.Share('3993.T')            #銘柄選択(例:日本株の場合=証券コード.T)
symbol_data = None
 
try:
    symbol_data = my_share.get_historical(
        share.PERIOD_TYPE_WEEK, 20,          #期間を選択
        share.FREQUENCY_TYPE_DAY, 1)        #足を選択(時間足、日足、週足、月足から選択可能)
except YahooFinanceError as e:
    print(e.message)
    sys.exit(1)
 
df = pd.DataFrame(symbol_data)              #DataFrameで二次元の表形式のデータに処理する※ここでsymbol_dataは辞書型になっている。イメージとしては辞書型のものを表にする感じ
df["datetime"] = pd.to_datetime(df.timestamp, unit="ms")    #datetimeカラムの追加。to_datetimeという関数はpandasのもの。unit="ms"はミリ秒まで表示。
#日本時間へ変換
df["datetime_JST"] = df["datetime"] + datetime.timedelta(hours=9)
df
```
