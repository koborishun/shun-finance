```python
import pandas
from pandas_datareader import data as pdr
import yfinance as yfin
yfin.pdr_override()

#a~bまでの調整済み株価を取得
df = pdr.get_data_yahoo(["3064.T", "3993.T", "4484.T", "5401.T", "5411.T", "9783.T"],start ='2022-01-01')
df = df["Adj Close"]
```

```python
df=df.rename(columns={'3064.T':'MRO','3993.T':'PKSHA','4484.T':'ランサーズ','5401.T':'日本製鉄','5411.T':'JFEHD','9783.T':'ベネッセHD',})
```

```python
df=df.fillna(method="ffill")#method='ffill'で直前の値が穴埋めの値に使われます。
df[df.index>="2022-01-01"]
```
